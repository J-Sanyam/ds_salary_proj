```python
def rating_bin(rating):
    if rating  == -1:
        return 'rating-1'
    elif rating  > -1 and rating <= 1:
        return 'rating1'
    elif rating  > 1 and rating <= 2:
        return 'rating2'
    elif rating > 2 and rating <= 3:
        return 'rating3'
    elif rating > 3 and rating <= 4:
        return 'rating4'
    else:
        return 'rating5'
```

```python
df['Rating'] = df['Rating'].apply(rating_bin)
```

```python
df['Company Age'] = df['Founded'].apply(lambda x : x if x == -1 else 2023 - x)
```

```python
df.drop(columns = ['Founded'], axis = 1, inplace  = True)
```

```python
age_bin = list(np.arange(df['Company Age'].drop_duplicates().nsmallest(2).iloc[-1], df['Company Age'].max() + 1, 20))
```

```python
age_bin.insert(0,-2)
```

```python
len(age_bin)
```

```python
age_labels = list(range(-1, 13, 1))
```

```python
df['Company Age'] = pd.cut(df['Company Age'], bins = age_bin, labels = age_labels, include_lowest = True)
```

```python
temp_df = list()
for i in df.index:
    temp_df.append('Age_' + str(df['Company Age'][i]))
    
df['Company Age'] = temp_df
```
